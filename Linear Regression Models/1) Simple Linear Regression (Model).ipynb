# 1) Simple Linear Regression (Model)

## Model

```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df = pd.read_csv("Data/Adv.csv")
df = df.iloc[:,1:len(df)]
df.head()
```

```python
df.info()
```

## Regresyon Grafiği İçin:

```python
sns.jointplot(x = "TV", y = "sales", data = df, kind = "reg", height=10);   # TV ile sales arasındaki regresyon grafiği
```

## Bağımlı Ve Bağımsız Değişkenlerin Tanımlanması

```python
x = df[["TV"]]

y = df[["sales"]]     #DataFrame Formatına Çevir Çünkü Regresyon Nesnesi Çalışmaz !!!
```

```python
y
```

## Regresyon Model Nesnesi:

```python
reg = LinearRegression()
```

```python
model = reg.fit(x,y)         # "...".fit("Bağımısz Değişken", "Bağımlı Değişken")
```

```python
model
```

```python
str(model)
```

```python
dir(model)
```

## Regresyon Değerlerini Almak İçin:

```python
model.intercept_  # B0 değeri 
```

```python
model.coef_    # B1 değeri
```

```python
## R-Kare Değeri

model.score(x,y)
```

## R-KARE: BAĞIMLI DEĞİŞKENDEKİ DEĞİŞİKLİĞİN, BAĞIMSIZ DEĞİŞKENLERCE AÇIKLAMA YÜZESİDİR.

```python
sns.jointplot(x = "radio", y = "sales", data =df, kind="reg",height=10);
```

```python
regresyon = LinearRegression()
```

```python
bağımlı_değişken = df[["sales"]]

bağımsız_değişken = df[["radio"]]
```

```python
bağımlı_değişken
```

```python
regresyon.fit(bağımsız_değişken,bağımlı_değişken)
```

```python
dir(regresyon)
```

```python
regresyon.coef_
```

```python
regresyon.score(bağımsız_değişken,bağımlı_değişken)  # Radio'nun satışlardaki değişikliği açıklama oranı yüzde 33'tür. 
```

# 2.Regresyon Deneme Projesi

```python
car = sns.load_dataset("mpg")
car = car.select_dtypes(include = ["int64","float64"])

car_copy = car.copy()
```

```python
car.columns
```

```python
car.isnull().sum()
```

```python
from ycimpute.imputer import knnimput
```

```python
car_değişkenleri = list(car)
```

```python
numpy_car = np.array(car)
```

```python
numpy_car
```

```python
tahmin_doldurması = knnimput.KNN(k=4).complete(numpy_car)
tahmin_doldurması
```

```python
car = pd.DataFrame(tahmin_doldurması, columns=car_değişkenleri)
car
```

```python
car.isnull().sum()
```

```python
sns.jointplot(x = "horsepower", y ="mpg", data =car, kind="reg", height=10)
```

```python
model_nesnesi = LinearRegression()
```

```python
bağımlı = car[["mpg"]]
bağımsız = car[["horsepower"]]
```

```python
regresyon_modeli = model_nesnesi.fit(bağımsız,bağımlı)
regresyon_modeli
```

```python
regresyon_modeli.coef_
```

```python
regresyon_modeli.intercept_
```

```python
regresyon_modeli.score(bağımsız,bağımlı)
```

# 3.Regresyon Deneme Projesi

```python
from ycimpute.imputer import EM
```

```python
pla = sns.load_dataset("planets")
pla = pla.select_dtypes(include = ["int64","float64"])
pla_copy = pla.copy()
```

```python
pla.dtypes
```

```python
pla.isnull().sum()
```

```python
pla_numpy = np.array(pla)
pla_columns = list(pla)
```

```python
eskik_veri_silme = EM().complete(pla_numpy)
eskik_veri_silme
```

```python
pla = pd.DataFrame(eskik_veri_silme, columns = pla_columns)
pla.isnull().sum()
```

```python
pla
```

```python
sns.jointplot(x = "mass", y= "orbital_period", data = pla, kind="reg", height=10)
```

```python
bgmlı = pla[["orbital_period"]]
bagmsız = pla[["mass"]]
```

```python
LinearRegression().fit(bagmsız,bgmlı)
```

```python
LinearRegression().fit(bagmsız,bgmlı).coef_
```

```python
LinearRegression().fit(bagmsız,bgmlı).intercept_
```

```python
LinearRegression().fit(bagmsız,bgmlı).score(bagmsız,bgmlı)
```

```python
LinearRegression().fit(bagmsız,bgmlı).score(bagmsız,bgmlı) * 100
```
